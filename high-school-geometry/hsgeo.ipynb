# Performing Transformations

A transformation is a function that changes the position, size or orientation of a shape. The three main types of transformations are:

* Translations
* Rotations
* Reflections

## Rigid Transformations

Rigid transformations are transformations that preserve shape and size of a figure. The main types of rigid transformations are:

* Translations
* Rotations
* Reflections

#### Translations

A translation is a transformation that moves a shape by a fixed distance in a given direction

```python
def translation_example():
    # Define the triangle with vertices A, B, and C
    triangle = np.array([[1, 1], [4, 1], [3, 4]])

    # Define the translation vector
    translation = np.array([3, 2])

    # Translate the triangle
    translated_triangle = np.add(triangle, translation)

    # Plot the original and translated triangles
    fig, ax = plt.subplots()
    ax.plot(*np.append(triangle, [triangle[0]], axis=0).T, marker='o', label='Original')
    ax.plot(*np.append(translated_triangle, [translated_triangle[0]], axis=0).T, marker='o', label='Translated')

    # Set labels and legend
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.legend()

    plt.show()

translation_example()
```

#### Rotations

A rotation is a transformation that turns a shape around a fixed point called the *center of rotation*.

```python
def rotation_example():
    def rotate(points, angle_degrees, center_of_rotation):
        angle_radians = np.radians(angle_degrees)
        rotation_matrix = np.array([[np.cos(angle_radians), -np.sin(angle_radians)],
                                    [np.sin(angle_radians), np.cos(angle_radians)]])
        translated_points = points - center_of_rotation
        rotated_translated_points = np.matmul(translated_points, rotation_matrix.T)
        return rotated_translated_points + center_of_rotation

    # Define the triangle with vertices A, B, and C
    triangle = np.array([[1, 1], [4, 1], [3, 4]])

    # Define the center of rotation and the angle
    center = np.array([2, 2])
    angle = 90

    # Rotate the triangle
    rotated_triangle = rotate(triangle, angle, center)

    # Plot the original and rotated triangles
    fig, ax = plt.subplots()
    ax.plot(*np.append(triangle, [triangle[0]], axis=0).T, marker='o', label='Original')
    ax.plot(*np.append(rotated_triangle, [rotated_triangle[0]], axis=0).T, marker='o', label='Rotated')

    # Set labels and legend
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.legend()

    plt.show()

rotation_example()
```
